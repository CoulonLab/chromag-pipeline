```python
from scipy import *
from matplotlib import pyplot as plt
import os
```

```python
basePath='C:/Users/Coulonlab/Desktop/Julia/analysis/'

lfn=['MAX_20200109_dCas9JF546-10-sg130g_1_MMStack_Pos0_cell1_s1_with-v1.2.1.trk2',
     'MAX_20200109_dCas9JF546-10-sg130g_1_MMStack_Pos0_cell1_s2_with-v1.2.1.trk2',
     'MAX_20200109_dCas9JF546-10-sg130g_1_MMStack_Pos0_cell2_s1_with-v1.2.1.trk2',
     'MAX_20200109_dCas9JF546-10-sg130g_1_MMStack_Pos0_cell2_s2_with-v1.2.1.trk2',
     'MAX_20200109_dCas9JF546-10-sg130g_1_MMStack_Pos0_cell3_s1_with-v1.2.1.trk2',
     ]

traces=array([loadtxt(basePath+fn) for fn in lfn])

lfn_bleach=['20200109_dCas9JF546-10-sg130g_1_MMStack_Pos1_cell1.csv',
            '20200109_dCas9JF546-10-sg130g_1_MMStack_Pos1_cell2.csv',
            ]

bleach_curves=array([loadtxt(basePath+fn,skiprows=1,delimiter=',') for fn in lfn_bleach])

bleach_curves[:,:,1]=(bleach_curves[:,:,1].T/bleach_curves[:,0,1]).T
mean_bleach_curve=bleach_curves.mean(0)
```

```python
fig=plt.figure(figsize=(12,8))

fig.add_subplot(221)
for i in range(traces.shape[0]):
    plt.plot(traces[i,:,0],traces[i,:,4]/1e3)
plt.xlabel('Time (frames)'); plt.ylabel('Spot fluo. (a.u.)');
#plt.yscale('log')

fig.add_subplot(222)
for i in range(traces.shape[0]):
    plt.plot(traces[i,:,0],traces[i,:,5]/1e3)
plt.xlabel('Time (frames)'); plt.ylabel('Spot background (a.u.)');
#plt.yscale('log')

fig.add_subplot(223)
for i in range(traces.shape[0]):
    plt.plot(traces[i,:,0],traces[i,:,4]/mean_bleach_curve[:,1]/1e3)
plt.xlabel('Time (frames)'); plt.ylabel('Spot fluo. (a.u.), bleach corrected');
#plt.yscale('log')

fig.add_subplot(224)
for i in range(bleach_curves.shape[0]):
    plt.plot(bleach_curves[i,:,0],bleach_curves[i,:,1])
plt.plot(mean_bleach_curve[:,0],mean_bleach_curve[:,1],c='k',lw=5)
plt.xlabel('Time (frames)'); plt.ylabel('Bleaching');
plt.ylim(0,None)
#plt.yscale('log')

plt.show()
```
